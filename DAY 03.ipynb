## INTRODUCTION

Visualization is the easiest way to analyze and absorb information because the way human brain processes information, using charts or graphs to visualize large amounts of complex data, is much easier than poring over spreadsheets or reports.

It is the first step towards any kind of data analysis work.

## MATPLOTLIB

## Pyplot

+ It is the state-machine interface to the plotting library in matplotlib.
+ In this figure and axes are implicitly and automatically created to achieve the desired plot. 
+ For Example, calling **plot()** function from pyplot module will automatically create the necessary figure and axes ro achieve the desired plot.
+ Setting a title will then automatically set that title to the current object axes. 
+ The pyplot interface is generally preferred for non-interactive plottinh (i.e, scripting)
+ Each pyplot function makes some changes to a figure, like creating a figure, creating a plotting area in a figure, plotting some lines in a plotting some lines in a plotting area, decorating the plot with labels etc.,

## LINE PLOT/CHART

````python
#simple line chart using two lists
# %matplotlib inline

import matplotlib.pyplot as plt
plt.plot([1,2,3], [5,7,4])  # plotting two lists
plt.show() # display the chart
````

````python
# program to draw two lines along with proper titles and legends
import matplotlib.pyplot as plt

x = [1,2,3]
y = [5,7,4]

plt.plot(x,y,label = "First Line") #plotting two lists with appropriate label

x2 = [1,2,3]
y2 = [10,11,14]
plt.plot(x2,y2, label = 'Second Line')

plt.xlabel("Plot number")
plt.ylabel("Important Variables")

plt.title("New Graph")
plt.legend()  #involves a legend

plt.show()  #displaying the chart
````

## Multiple Views

````python
# plotting line chart in different views
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0.0, 20.0, 1)
s = list(range(1,21))
s2 = list(range(4,24))

plt.subplot(2,1,1) # the figure has 2 row, 1 columns, and this plot is the first plot. 
plt.plot(t,s)
plt.ylabel("Values")
plt.title("First Chart")
plt.grid(True)

plt.subplot(2,1,2) # the figure has 2 row, 1 columns, and this plot is the second plot.
plt.plot(t,s2)
plt.xlabel("Item (s)")
plt.ylabel('Value')
plt.title("\n\n Second Chart")
plt.grid(True)

plt.show()
````

````python
# plotting line chart in different views
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0.0, 20.0, 1)
s = list(range(1,21))
s2 = list(range(4,24))

plt.subplot(1,2,1) #the figure has 1 row, 2 columns, and this plot is the first plot. 
plt.plot(t,s)
plt.ylabel("Values")
plt.title("First Chart")
plt.grid(True)

plt.subplot(1,2,2) #the figure has 1 row, 2 columns, and this plot is the second plot. 
plt.plot(t,s2)
plt.xlabel("Item (s)")
plt.ylabel('Value')
plt.title("\n\n Second Chart")
plt.grid(True)

plt.show()
````

### Adjusting the space between the subplots

````python
# plotting line chart in different views
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0.0, 20.0, 1)
s = list(range(1,21))
s2 = list(range(4,24))

plt.subplot(2,1,1) # the figure has 2 row, 1 columns, and this plot is the first plot. 
plt.plot(t,s)
plt.ylabel("Values")
plt.title("First Chart")
plt.grid(True)

plt.subplots_adjust(hspace = 0.5, wspace = 0.5)  # ADJUST the Spacing between

plt.subplot(2,1,2) # the figure has 2 row, 1 columns, and this plot is the second plot.
plt.plot(t,s2)
plt.xlabel("Item (s)")
plt.ylabel('Value')
plt.title("\n\n Second Chart")
plt.grid(True)

plt.show()
````

````python
#Program to plot frequency of marks using line chart

import matplotlib.pyplot as plt

def fnplot(list):
    plt.plot(list1)
    plt.title("Marks Line Chart")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.show()
    
list1 = [50,50,50,65,65,75,75,80,80,90,90,90]
fnplot(list1)
````

````python
#program for plotting sine wave using line chart
import matplotlib.pyplot as plt
import numpy as np

xvals = np.arange(-10,10,0.01)
yvals = np.sin(xvals)
plt.plot(xvals, yvals)
plt.show()
````

````python
#program to evaluate an algebraic expression 10x+14 using line chart

import numpy as np
from matplotlib import pyplot as plt

x = np.arange(12,20)
y = 10*x + 14

plt.title("Graph for an Algebraic Expression")
plt.xlabel("x axis")
plt.ylabel("y label")

plt.plot(x,y)
plt.show()
````

````python
# Program to plot a quadratic equation using dashed line chart
# 1 - 0.5*x**2

import numpy as np
from matplotlib import pyplot as plt

xvals = np.arange(-2, 1, 0.01)
newyvals = 1 - 0.5* xvals **2

plt.plot(xvals, newyvals, 'b--')
plt.title("Example plots")
plt.xlabel("Input")
plt.ylabel("Function Values")
plt.show()
````

# PIE PLOT/CHART

````python
import matplotlib.pyplot as plt

#Data to plot
labels = ['Python', 'C++', 'Ruby', 'Java']
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)

plt.pie(sizes, explode = explode, labels = labels, colors = colors, shadow = True, startangle = 140)

plt.axis('equal')
plt.show()
````

# SCATTER CHART

````python
import matplotlib.pyplot as plt

weight = [83.3, 74, 64.5, 79, 90, 84.8, 84, 88,66,77, 79, 90.5, 74.3, 90, 73]
height = [167, 113.7, 124.8, 165.5, 145.4, 155, 149, 180, 169.5, 156, 167.3, 145, 185, 181, 168.3]

plt.scatter(weight, height)

plt.xlabel('weight')
plt.ylabel('height')

plt.title('Scatter plot-height vs weight', fontsize = 20)
plt.show()
````

````python
import matplotlib.pyplot as plt

weight = [83.3, 74, 64.5, 79, 90, 84.8, 84, 88,66,77, 79, 90.5, 74.3, 90, 73]
height = [167, 113.7, 124.8, 165.5, 145.4, 155, 149, 180, 169.5, 156, 167.3, 145, 185, 181, 168.3]

plt.scatter(weight, height, c = 'r', marker = '*')

plt.xlabel('weight', fontsize = 16)
plt.ylabel('height', fontsize = 16)

plt.title('Scatter plot-height vs weight', fontsize = 20)
plt.show()
````

# BAR PLOT/CHART

````python
#To plot a simple bar chart

import matplotlib.pyplot as plt

y_axis = [20,50,30]
x_axis = range(len(y_axis))

plt.bar(x_axis, y_axis, width = 0.5, color = 'orange')
plt.show()
````

````python
#to plot two list elements using bar chart

import matplotlib.pyplot as plt

x = [2,4,6,8,10]
y = [6,7,8,2,4]

x2 = [1,3,5,7,9]
y2 = [7,8,2,4,2]

plt.bar(x,y, label = "Bars1")
plt.bar(x2, y2, label = "bars2")

plt.xlabel('x')
plt.ylabel('y')

plt.title("Bar Graph \n With multiline title")
plt.legend()
plt.show()
````

````python
#program to plot a bar chart showing the choice of favourite movie among the people

import numpy as np
import matplotlib.pyplot as plt

objects = {'Comedy', 'Action', 'Romance', 'Drama', 'SciFi'}
y_pos = np.arange(len(objects))

types = (4,5,6,1,4)
plt.bar(y_pos, types, align = 'center', color = 'blue')

plt.xticks(y_pos, objects)

plt.ylabel("People")
plt.title("Favourite Type of Movie")
plt.show()
````

````python
#to plot a histogram for number of students with their respective weights

import matplotlib.pyplot as plt

data_students = [5,15,25,35,45,55]
plt.hist(data_students, bins = [0,10,20,30,40,50,60], weights = [20,10,45,33,6,8], edgecolor = "red")
plt.show()
````

````python
#to plot a histogram with different formatting and suitable title and labels

import numpy as np
import matplotlib.pyplot as plt

data_students = [1,11,21,31,41,51]

plt.hist(data_students, 
         bins = [0,10,20,30,40,50,60], 
         weights = [10, 1, 0, 33, 6,8], 
         facecolor = 'y', 
         edgecolor = 'red'
        )

plt.title("Histogram for students data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.savefig("student.png")
plt.show()
````

## FREQUENCY PLOYGONS

````python
#to construct a frequency polygon on the basis of student_data v/s weight histogram

import numpy as np
import matplotlib.pyplot as plt

data = [5,15,25,35,15,55]
plt.hist(data, bins = [0,10,20,30,40,50,60], weights = [20,10,45,33,6,6],
        edgecolor = 'red', histtype = 'step')


plt.xlabel("Value")
plt.ylabel("Probability")
plt.title("Histogram")
plt.show()
````
